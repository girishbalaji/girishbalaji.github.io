## Traffic

```python
# 5mm
traffic_5mm = plt.imread(pinhole_data + "IMG_0289.JPG")
traffic_5mm = np.flipud(traffic_5mm)
traffic_5mm = traffic_5mm[1300:3400,1200:3400]
traffic_5mm = exposure.equalize_hist(traffic_5mm)

blurred = traffic_5mm.copy().astype(float)

# Denoise the Image
blurred = restoration.denoise_wavelet(blurred, multichannel=True)
blurred = restoration.denoise_bilateral(blurred, multichannel=True)

# Sharpen the image

# Laplacian Kernel
kernel      = np.ones((3,3,3))*(-1)
kernel[1,1,:] = 8

#Convolution of the image with the kernel:
lap_im = ndimage.filters.convolve(blurred, kernel)

k = 4 #Sharpening factor!
final_lap = lap_im*k/np.amax(lap_im) 

# Add negative Laplacian to the original image:
sharpened = blurred - final_lap 

# Set negative values to 0, values over 255 to 255:
sharpened = np.clip(sharpened, 0, 1)

plt.imshow(sharpened)
plt.show()

plt.imsave(output_data + "best-traffic.jpg", sharpened, format = "jpg")
```
